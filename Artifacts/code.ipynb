EVALUATION

```python
def Evaluation(x,y):
    try:
        # print("R2 score of the model is:",round(r2_score(x,y)*100,2),"%")
        # print("MAE of the model is:",mean_absolute_error(x,y))
        # print("MSE of the model is:",mean_squared_error(x,y))
        # print("RMSE of the model is:",np.sqrt(mean_squared_error(x,y)))
        return round(r2_score(x,y)*100,2), round(mean_absolute_error(x,y),2), round(mean_squared_error(x,y),2), round(np.sqrt(mean_squared_error(x,y)),2)
    except:
        print("Error in Evaluation!")
```

```python
def ResultinDataframe(y_pred, y_pred_KNN,y_pred_dt,y_pred_rf,y_pred_ad):
    Result = pd.DataFrame({"Model":["Linear Regression","KNN","DT","RF","ADA"],
                           "R2 Score":None,"MAE":None,"MSE":None,"RMSE":None})
    
    #Impute linear regression values in dataframe
    a,b,c,d = Evaluation(y_test,y_pred)
    Result.iloc[0,1] = a
    Result.iloc[0,2] = b
    Result.iloc[0,3] = c
    Result.iloc[0,4] = d

    #Impute KNN values in dataframe
    a,b,c,d = Evaluation(y_test,y_pred_KNN)
    Result.iloc[1,1] = a
    Result.iloc[1,2] = b
    Result.iloc[1,3] = c
    Result.iloc[1,4] = d   

    #Impute DT values in dataframe
    a,b,c,d = Evaluation(y_test,y_pred_dt)
    Result.iloc[2,1] = a
    Result.iloc[2,2] = b
    Result.iloc[2,3] = c
    Result.iloc[2,4] = d

    #Impute RF values in dataframe
    a,b,c,d = Evaluation(y_test,y_pred_rf)
    Result.iloc[3,1] = a
    Result.iloc[3,2] = b
    Result.iloc[3,3] = c
    Result.iloc[3,4] = d

    #Impute ADA values in dataframe
    a,b,c,d = Evaluation(y_test,y_pred_ad)
    Result.iloc[4,1] = a
    Result.iloc[4,2] = b
    Result.iloc[4,3] = c
    Result.iloc[4,4] = d

    return Result
```

```python
Finaldf = ResultinDataframe(y_pred, y_pred_KNN,y_pred_dt,y_pred_rf,y_pred_ad)
Finaldf
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import RandomizedSearchCV

# Define the KNN Regressor
knn = KNeighborsRegressor()

# Define the parameter grid for the randomized search
param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2],
}

# Perform randomized search with cross-validation
random_search = RandomizedSearchCV(estimator=knn, param_distributions=param_grid, n_iter=10, cv=5, random_state=42)
random_search.fit(x_train, y_train)

# Print the best parameters
print("Best parameters:", random_search.best_params_)

# Create a new KNN Regressor with the best parameters
best_knn = KNeighborsRegressor(**random_search.best_params_)

# Train the model
best_knn.fit(x_train, y_train)

# Make predictions on the training set
new_y_pred_knn = best_knn.predict(x_test)

# Evaluate the model
Evaluation(y_test, new_y_pred_knn)
```
